## Helper Method

````python
import re
from typing import List

def regex_helper(pattern: str, candidates: List[str]):
  match_candidates: List[str] = []
  skip_candidates: List[str] = []

  for candidate in candidates:
    is_match = re.search(pattern, candidate)
    if is_match:
      match_candidates.append(is_match)
    else:
      skip_candidates.append(candidate)

  print("The results are as follows")
  print("Match: ", ", ".join([item.group() for item in match_candidates]))
  print("Skip : ", ", ".join([item for item in skip_candidates]))
  return match_candidates, skip_candidates
````

### Contoh #1

Buatlah *regex pattern* sedemikian sehingga memenuhi kondisi di bawah ini
```
Task  | Text
Match | abcd!
Match | abdd!
Match | abbccddd!
Match | aabcccddd!

Skip  | bcd!
Skip  | acd!
Skip  | abc!
```
Anda diharapkan mampu membangun *regex pattern* untuk menangkap pola berstatus **Match** serta melewati/tidak menangkap pola berstatus **Skip**.

Apabila dilihat dari pola regex di atas, Anda diharapkan untuk membangun pola regex yang wajib terdiri dari huruf a,b,d secara urut, dengan karakter 'c' yang bersifat opsional, lalu diikuti dengan tanda baca "!".


````python
pattern = r'a+b+c*d+!'
candidates = ['abcd!', 'abdd!', 'abbccddd!', 'aabcccddd!', 'bcd!', 'acd!', 'abc!']
match_candidates, skip_candidates = regex_helper(pattern, candidates)

print(match_candidates)
print(skip_candidates)
````
